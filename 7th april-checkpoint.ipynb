```python
target_name="y"
y=df[target_name]
x=df.drop(target_name,axis=1)
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
ssx=scaler.transform(x)
```

```python
#TRAINING THE MODEL
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(ssx,y,test_size=0.3, random_state=7)
```

```python
y_train.shape,x_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from xgboost import XGBClassifier # this isfor data training
from sklearn.metrics import accuracy_score
model=XGBClassifier()
model.fit(x_train,y_train)
print(model)
```

```python
#make predictions for the data 
predict=model.predict(x_test)
predictions=[round(value)for value in predict]
```

```python
#Evaluating the prediction from the test
#accuracy=accuracy_score(predict,predictions)
#print("accuracy: %.2f%%" %(accuracy*100))
```

```python
print("Accuracy of model::",accuracy_score(y_test,predictions)*100,"%")
```
